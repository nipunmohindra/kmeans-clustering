## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
# Load the data
raw_data = pd.read_csv(r'D:\Stu\Data Science\Exercises Regression\Countries_exercise.csv')
# Check the data
raw_data
```

## Plot the data

```python
plt.scatter(raw_data['Longitude'], raw_data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

## Select the features

```python
x = raw_data.iloc[:,1:3]
x
```

## Clustering

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(x)
```

### Clustering Resutls

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = raw_data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(raw_data['Longitude'], raw_data['Latitude'], c=data_with_clusters['Cluster'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

## Selecting the number of clusters

### WCSS

```python
kmeans.inertia_
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
wcss = []
# 'cl_num' is a that keeps track the highest number of clusters we want to use the WCSS method for.
# Note that 'range' doesn't include the upper boundery
cl_num = 11
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

### The Elbow Method

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

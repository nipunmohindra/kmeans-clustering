```python
#Defining eucledian for distance calculation
def euclidean(p,q):
    d = 0

    for i in range(len(p)):
        d = d + (p[i]-q[i])**2

    return(np.sqrt(d))
```

```python
#finding the distance  
dist = []

for i in range(len(trainX)):
    dist.append(euclidean(testX, trainX.iloc[i,[0,1,2]])) #taking0,1,2 because we have trainX and trainY as well
    
print(dist)
```

```python
#Now define the knn function
def knn(trainX,trainY,testX,k): 
    
    
    dist = []
    

    for i in range(0, len(trainX)):
        dist.append(euclidean(testX,trainX.iloc[i,[0,1,2]]))

    d = dist.copy()
    dist.sort()
    #print(d)

    index = []
    for i in range(0,k):
        indx = d.index(dist[i])
        index.append(indx)
        d[indx] = -1
        #print(d)
        
    #print(index)
    

    pred_y = np.mean(trainY.iloc[index])
    return(pred_y)
```

```python
#Running knn for k=1
knn(trainX,trainY,testX,1)
```
